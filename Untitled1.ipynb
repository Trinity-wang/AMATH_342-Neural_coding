```python
import numpy as np
import matplotlib.pylab as plt

dt = 0.001
times = np.arange(0, 1 + dt, dt)
yVals = np.exp(np.sin(times))

plt.plot(times, yVals)
plt.title("Graphing e^(sin(t)) on [0,1]")
plt.grid(axis = 'both')
```

```python
import scipy.integrate as scp

def RCeqn(y,t):
    R = 1
    C = 1
    vPrime = [-1*y[0] / (R*C)]
    
    return vPrime

dt = 0.001
v0 = 1
times = np.arange(0, 2 + dt, dt)
Vt = scp.odeint(RCeqn, v0, times)
plt.plot(times, Vt)
plt.grid(axis = 'both')
plt.title("Numerically solving RC circuit Equation")
plt.ylabel("V(t)")
plt.xlabel("time")
```
